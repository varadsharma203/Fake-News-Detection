#STEAMMING
steamming is the process of reducing a word to its root word likke removiing the prefix and sufixes of different words

we will be using porter streamming for this

```python
port_stem= PorterStemmer()
```

```python
def stemming(content):
  stemmed_content=re.sub('[^a-zA-Z]',' ',content)
  stemmed_content=stemmed_content.lower()
  stemmed_content=stemmed_content.split()
  stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content=' '.join(stemmed_content)
  return stemmed_content
```

```python
#now we will apply this function to our coloums
news_dataset['content']=news_dataset['content'].apply(stemming)
```

```python
print(news_dataset['content'])
```

```python
#seperating the data and lables
X=news_dataset['content'].values
Y=news_dataset['label'].values
```

```python
print(X)
print(Y)
```

```python
#seperating the data and lables
X=news_dataset['content'].values
#converting textual data to numerical data
vectorizer=TfidfVectorizer()
X = vectorizer.fit_transform(X)
print(X)
```

```python
#splitting the data into  training and test data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, stratify=Y , random_state=2)
```

```python
#training the model
model=LogisticRegression()
model.fit(X_train,Y_train)
```

Evaluation
accuracy score

```python
#accuracy score of the training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print("accuracy score of the training data",training_data_accuracy)
```

```python
#accuracy score of the test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print("accuracy score of the test data",test_data_accuracy)
```
